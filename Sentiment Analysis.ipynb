```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import requests
from bs4 import BeautifulSoup
import re
```

```python
### Instantiate and set up the model
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
### Encoding and Calculating sentiment
tokens = tokenizer.encode('It was ok. But could be better', return_tensors = 'pt')
```

```python
### converting comments to a sequence of numbers
result = model(tokens)
```

```python
result.logits
```

```python
### giving integer value of 1-5 of comments
int(torch.argmax(result.logits)) + 1
```

```python
r = requests.get('https://www.imdb.com/title/tt9362722/reviews?ref_=tt_urv')
soup = BeautifulSoup(r.text, 'html.parser')
regex = re.compile('.*text show-more__control*.')
results = soup.find_all('div', {'class':regex})
reviews = [result.text for result in results]
```

```python
reviews
```

```python
### loading the reviews in dataframe and scoring them
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.array(reviews), columns = ['review'])
```

```python
df['review'].iloc[0]
```

```python
def sentiment_score(review):
    tokens = tokenizer.encode(review, return_tensors = 'pt')
    result = model(tokens)
    return int(torch.argmax(result.logits)) + 1
```

```python
sentiment_score(df['review'].iloc[1])
```

```python
df['sentiment'] = df['review'].apply(lambda x: sentiment_score(x[:512]))
```
